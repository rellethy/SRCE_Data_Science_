```python
# prompt: calculate the total number of entries for 2021 2022 and 2023 separately

import pandas as pd

netflix = pd.read_csv('/content/Netflix Userbase.csv')

# Convert 'Join Date' to datetime objects
netflix['Join Date'] = pd.to_datetime(netflix['Join Date'], format='%d-%m-%y')

# Extract year from 'Join Date'
netflix['Year'] = netflix['Join Date'].dt.year

# Calculate the total number of entries for each year
entries_2021 = len(netflix[netflix['Year'] == 2021])
entries_2022 = len(netflix[netflix['Year'] == 2022])
entries_2023 = len(netflix[netflix['Year'] == 2023])

print(f"Total entries for 2021: {entries_2021}")
print(f"Total entries for 2022: {entries_2022}")
print(f"Total entries for 2023: {entries_2023}")
```

```python
# prompt: find the number of entries before nov 1 2022 and after

# Assuming 'netflix' DataFrame is already loaded and 'Join Date' is in datetime format

# Filter entries before Nov 1, 2022
before_nov_2022 = netflix[netflix['Join Date'] < pd.to_datetime('2022-11-01')]
num_before_nov_2022 = len(before_nov_2022)

# Filter entries after Nov 1, 2022
after_nov_2022 = netflix[netflix['Join Date'] >= pd.to_datetime('2022-11-01')]
num_after_nov_2022 = len(after_nov_2022)


print(f"Number of entries before Nov 1, 2022: {num_before_nov_2022}")
print(f"Number of entries after Nov 1, 2022: {num_after_nov_2022}")
```
