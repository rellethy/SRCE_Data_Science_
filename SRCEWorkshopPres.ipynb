```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

data = pd.read_csv('/content/cleaned_star_data.csv')

data

data.head()
```

```python
predictors = ['Luminosity(L/Lo)', 'Absolute magnitude(Mv)', 'Radius(R/Ro)', 'Star type', 'Star color', 'Spectral Class']
for column in predictors:
  plt.figure()
  sns.histplot(data[column], kde=True)
  plt.title(f'Distribution of {column}')
  plt.xlabel(column)
  plt.show()

```

```python
star_color_mapping = {'Red':1, 'Blue':2, 'White': 3, 'Blue-White': 4, 'Yellow-White' : 5}

data['Star color'] = data['Star color'].map(star_color_mapping)

data
```

```python
spectral_class_mapping = {'M' : 1, 'K': 2, 'G': 3, 'F' : 4, 'A' : 5, 'B' : 6, 'O' : 7}

data['Spectral Class'] = data['Spectral Class'].map(spectral_class_mapping)

data
```

```python
correlation_matrix = data.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

```python
import statsmodels.api as sm


y = data['Temperature (K)']
X = data.drop('Temperature (K)', axis=1)

X = sm.add_constant(X)

model = sm.OLS(y, X).fit()

print(model.summary())
```
